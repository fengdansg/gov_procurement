```python
df = pd.read_csv('government-procurement-via-gebiz.csv')
df.head()
```

```python
text = " ".join(desc for desc in df.tender_description)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["Delivery","Supply","Contract","Term","YEAR","Tender","OPTION","EXTEND","TWO YEARS","System","Invitation", "Provision","TWO", "YEARS","THREE","ANOTHER","AGREEMENT","one","unit","related","UNIVERSITY","SINGAPORE","REQUEST","PROPOSAL","Period","Systems","APPOINTMENT"])

# Generate a word cloud image
wordcloud = WordCloud(max_font_size=80, scale=3,stopwords=stopwords, max_words=100,  background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

# Testing for one description

```python
nlp = spacy.load('en')
from spacy.lang.en.stop_words import STOP_WORDS
len(STOP_WORDS)
```

```python
text=nlp(df.tender_description[1])
t_l = ""
for token in text:
    t_l +=token.lower_+" "
    

print(t_l)
    
```

```python
customize_stop_words = [
    'provision', 'acra','invitation'
]
for w in customize_stop_words:
    nlp.vocab[w].is_stop = True

text_rm=nlp(t_l)
category=""
previous_item=""
other=""

for n in text_rm.noun_chunks:
    check_noun=""
    n_nlp= nlp(n.text)
    for token in n_nlp:
        if not token.is_stop:
            check_noun+= token.text+" "
    print("checked noun is: "+ check_noun)
    if "service" in n.text:
        if "related service" in n.text :
            category= previous_item 
           
        else:
            category=n.text
   
    elif "construction" in n.text:
            category=n.text
    elif "building" in n.text:
            category=n.text
    else:
            other +=n.text+" "
    previous_item=n.text;

if category=="":
    category=other
    
print(category)
print(other)
```
